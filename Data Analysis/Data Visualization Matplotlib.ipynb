##Which one to use ? pyplot vs matplotlib OO method?
for quickly=pyplot
for advanced = matplotlib

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50[:10]
```

```python
##Pyplot METHOD

over_50.plot(kind="scatter", x='age',y='chol',c='target');
```

```python
#00 METHOD MIXED WITH PYPLOT
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x='age',y='chol',c='target',ax=ax);
```

```python
 #OO METHOD FROM SCRATCH
fix,ax = plt.subplots(figsize=(10,6))

#plotting
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

#customizing
ax.set(title="Heart Disease And Cholesterol levels ",
       xlabel="Age",
       ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(),title="Target");
##looks in the scatter variables and looks for legend element from c and finds the target##


#Add a horizontal line

ax.axhline(over_50["chol"].mean(),linestyle="--")
```

```python
##Subplot of chol,age,thalch
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                          sharex=True)

#add data to ax0

scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])
   #Customize ax0

ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")
   #Add legend
ax0.legend(*scatter.legend_elements(),title="Target")
   #Add mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

#add data to ax1

scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])

#Customize ax1

ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate ")
   #Add legend
ax1.legend(*scatter.legend_elements(),title="Target")
   #Add mean line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");


#Add a title to Figure in General

fig.suptitle("Heart Disease Analysis",fontsize=18, fontweight="bold")
```

```python

#Set Stylle
plt.style.use('seaborn')

#OO METHOD FROM SCRATCH
fix,ax = plt.subplots(figsize=(10,6))

#plotting
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")#this changes color scheme

#customizing
ax.set(title="Heart Disease And Cholesterol levels ",
       xlabel="Age",
       ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(),title="Target");
##looks in the scatter variables and looks for legend element from c and finds the target##


#Add a horizontal line

ax.axhline(over_50["chol"].mean(),linestyle="--")
```

```python
#CUSTOMIZING X AND Y LIMITATIONS


##Subplot of chol,age,thalch
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                          sharex=True)

#add data to ax0

scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")
   #Customize ax0

ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

    #cHANGE ax0 x axis limit
ax0.set_xlim(50,80)
   #Add legend
ax0.legend(*scatter.legend_elements(),title="Target")
   #Add mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

#add data to ax1

scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter")

#Customize ax1

ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate ")

   #cHANGE ax0 x axis limit
ax1.set_xlim(50,80)
ax1.set_ylim(60,200)
   #Add legend
ax1.legend(*scatter.legend_elements(),title="Target")
   #Add mean line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");


#Add a title to Figure in General

fig.suptitle("Heart Disease Analysis",fontsize=18, fontweight="bold")
```
